```python
import pandas as pd
```

```python
# Ejercicio 1: apertura y carga de datos en un dataframe
df = pd.read_csv("Entrenamieto_ECI_2020.csv", encoding="latin-1")
```

```python
df.head()
```

```python
# Me toma la primera columna con un titulo raro, lo corrijo
df.rename(columns={'ï»¿ID' : 'ID'}, inplace=True)
```

```python
df.head()
```

```python
# Ejercicio 2: mostrar para la region EMEA los datos ID, Territory, Delivery_Quarter, Delivery_Year para los 10 proyectos
# con mayor Total Amount

df.loc[df['Region']=='EMEA', ['ID', 'Territory', 'Delivery_Quarter', 'Delivery_Year', 'Total_Amount']]\
    .nlargest(10, 'Total_Amount')
```

```python
# Ejercicio 3: calcular el promedio del precio (Price) para el territorio Germany

df_germany_price = df.loc[df['Territory']=='Germany', ['Price']]
df_germany_price.head()
```

```python
# Hay valores no numericos: "None" y "Other"
df_germany_price['Price'].value_counts()
```

```python
germany_total = len(df_germany_price)
print("Total de registros para territorio Germany: ", germany_total)
```

```python
# Cambio el tipo de dato a numerico para poder calcular el promedio, casteo strings a NaN y los remuevo
df_germany_price['Price'] = pd.to_numeric(df_germany_price['Price'], errors='coerce')
df_germany_price.dropna(inplace=True)
df_germany_price.head()
```

```python
# Verifico el cambio en el tipo de datos
df_germany_price.dtypes
```

```python
# Se elimina la mayoria de los registros. 
# No se analizo si es lo mas conveniente removerlos, reemplazarlos por el promedio, la mediana o usar otro metodo

germany_with_price = len(df_germany_price)
print("Total de registros para el territorio Germany con Price numerico: ", germany_with_price)
print("Registros que se quedaron: ", str(round(((germany_with_price/germany_total)*100), 2)) + "%")
```

```python
# Calculo el promedio

avg = df_germany_price['Price'].sum()/germany_with_price
avg
```
